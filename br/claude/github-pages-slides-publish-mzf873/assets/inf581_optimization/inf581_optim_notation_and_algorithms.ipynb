An MDP is a Tuple $(\statespacefull, \actionspace, \mathrm{P}_t, r_t, \gamma)$ such as:
- $\statespacefull$ is the state space;
- $\actionspace$ is the action space;
- $T$ is the time axis;
- $\transitionfunc \in(P_t)_{t \in T}$ is a family of markovian transition probability distributions between states conditionned on actions;
- $(r_t)_{t \in T}$ is a bouded family of rewards associated to transitions
- $\gamma$ is a discount factor

## Differential Evolution

**Input**:<br>
$\quad\quad$ $f$: the objective function<br>

**Algorithm parameter**:<br>
$\quad\quad$ ...

Arbitrarily initialize the population<br>

**FOR EACH** generation<br>
$\quad$ **FOR EACH** children<br>
$\quad\quad$ 1. *MUTATION*<br>
$\quad\quad\quad$ Choose 3 random distinct individuals $\boldsymbol{a}$, $\boldsymbol{b}$ et $\boldsymbol{c}$<br>
$\quad\quad\quad$ $\boldsymbol{b}$ and $\boldsymbol{c}$ define the direction and the amplitude of the mutation (i.e. the search vector)<br>
<br>
$\quad\quad\quad$ Combine them to make an interim individual $\boldsymbol{z} \leftarrow \boldsymbol{a} + \omega ~ (\boldsymbol{b} - \boldsymbol{c})$<br>
$\quad\quad\quad$ where $\omega$ is a mutation amplification factor (a scalar)<br>
<br>
$\quad\quad$ 2. *CROSSOVER*<br>
$\quad\quad\quad$ Choose a random dimension $j \in [1,\dots,n]$ for optimization in $n$ dimensions<br>
$\quad\quad\quad$ Make the candidate individual $\boldsymbol{x}'$ using the binary crossover<br>
<br>
<br>
<br>
$\quad\quad$ 3. *SELECTION*<br>
$\quad\quad\quad$ Insert the best individual between $\boldsymbol{x}'$ and $\boldsymbol{x}$ into the next generation<br>

**RETURN** the best individual in the last generation

## Generic EA

**Input**:<br>
$\quad\quad$ $f$: the objective function<br>

**Algorithm parameter**:<br>
$\quad\quad$ $\mu$: number of parents per generation<br>
$\quad\quad$ $\rho$: number of parents used to make each child<br>
$\quad\quad$ $\lambda$: number of children per generation<br>
$\quad\quad$ mutation operator<br>
$\quad\quad$ mating selection operator<br>
$\quad\quad$ survivor selection operator<br>
$\quad\quad$ termination criterion<br>
$\quad\quad$ an initialization procedure<br>

Initialize a population of $\mu$ individuals<br>
Evaluate each individual $\boldsymbol{x}$: $f(\boldsymbol{x})$<br>

**FOR EACH** generation (until termination criterion)<br>
$\quad$ **FOR EACH** child $\boldsymbol{x}$<br>
$\quad\quad$ 1. select $\rho$ parents<br>
$\quad\quad$ 2. $\boldsymbol{x} \leftarrow$ recombinate selected parents (crossover)<br>
$\quad\quad$ 3. mutate $\boldsymbol{x}$<br>
$\quad\quad$ 4. evaluate $f(\boldsymbol{x})$<br>
$\quad$ Select the next generation of $\mu$ individuals<br>

**RETURN** the best individual

## SAES: $(\mu/\rho, \lambda)-\sigma-SA-ES$ and $(\mu/\rho + \lambda)-\sigma-SA-ES$

**Input**:<br>
$\quad\quad$ $f$: the objective function<br>

**Algorithm parameter**:<br>
$\quad\quad$ $\tau$: self-adaptation learning rate<br>
$\quad\quad$ $\mu$: number of parents per generation<br>
$\quad\quad$ $\rho$: number of parents used to make each child<br>
$\quad\quad$ $\lambda$: number of children per generation<br>
$\quad\quad$ mating selection operator<br>
$\quad\quad$ survivor selection operator<br>

Initialize arbitrarily the population<br>

**FOR EACH** generation<br>
$\quad$ **FOR EACH** children<br>
$\quad\quad$ 1. randomly select $\rho$ parents<br>
$\quad\quad$ 2. $(\boldsymbol{x}, \sigma) \leftarrow$ recombination of selected parents (if $\rho > 1$)<br>
$\quad\quad$ 3. mutation of $\sigma$ (individual strategy) : $\sigma \leftarrow \sigma ~ e^{\tau \mathcal{N}(0,1)}$<br>
$\quad\quad$ 4. mutation of $\boldsymbol{x}$ (objective param) : $\boldsymbol{x} \leftarrow \boldsymbol{x} + \sigma ~ \mathcal{N}(0,1)$<br>
$\quad\quad$ 5. evaluate $f(\boldsymbol{x})$<br>
$\quad$ Select the next generation of individuals<br>

**RETURN** the best individual

## (1+1)-SA-ES

**Input**:<br>
$\quad\quad$ $f$: the objective function<br>
$\quad\quad$ $\boldsymbol{x}$: initial solution<br>

**Algorithm parameter**:<br>
$\quad\quad$ $\tau$: self-adaptation learning rate<br>

**FOR EACH** generation<br>
$\quad\quad$ 1. mutation of $\sigma$ (current individual strategy) : $\sigma' \leftarrow \sigma ~ e^{\tau \mathcal{N}(0,1)}$<br>
$\quad\quad$ 2. mutation of $\boldsymbol{x}$ (current solution) : $\boldsymbol{x}' \leftarrow \boldsymbol{x} + \sigma' ~ \mathcal{N}(0,1)$<br>
$\quad\quad$ 3. eval $f(\boldsymbol{x}')$<br>
$\quad\quad$ 4. survivor selection $\boldsymbol{x} \leftarrow \boldsymbol{x}'$ and $\sigma \leftarrow \sigma'$ if $f(\boldsymbol{x}') \leq f(\boldsymbol{x})$<br>

**RETURN** $\boldsymbol{x}$

## Cross Entropy

**Input**:<br>
$\quad\quad$ $f$: the objective function<br>
$\quad\quad$ $\mathbb{P}$: family of distribution<br>
$\quad\quad$ $\boldsymbol{\theta}$: initial parameters for the proposal distribution $\mathbb{P}$<br>

**Algorithm parameter**:<br>
$\quad\quad$ $m$: sample size<br>
$\quad\quad$ $m_{\text{elite}}$: number of samples to use to fit $\boldsymbol{\theta}$<br>

**FOR EACH** iteration<br>
$\quad\quad$ samples $\leftarrow \{ \boldsymbol{x}_1, \dots, \boldsymbol{x}_m \}$ with $\boldsymbol{x}_i \sim \mathbb{P}(\boldsymbol{\theta}) ~~ \forall i \in 1\dots m$<br>
$\quad\quad$ elite $\leftarrow $ { $m_{\text{elite}}$ best samples } $\quad$ (i.e. select best samples according to $f$)<br>
$\quad\quad$ $\boldsymbol{\theta} \leftarrow $ fit $\mathbb{P}(\boldsymbol{\theta})$ to the elite samples<br>

**RETURN** $\boldsymbol{\theta}$
